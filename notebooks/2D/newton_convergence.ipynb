```python
import plotly.graph_objects as go
import pandas
import numpy
```

```python
df = pandas.read_csv("energy-convergence.csv")
df = pandas.read_csv("energy-convergence-sqrt.csv")
display(df)
for name, data in df.iteritems():
    if(name != "Iterations"):
        df[name] = numpy.log10(data)
```

```python
fig = go.Figure(data=[go.Scatter(y=y, name=name) for name, y in df.iteritems() if name != "Iterations"])
fig.update_layout(
    title="Converged Values",
    xaxis_title="time-step",
    yaxis_title="$\log_{10}(\\text{criteria})$",
)
fig.show()
```
