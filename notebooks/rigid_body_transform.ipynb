# Rigid Body Transformations

```python
import numpy as np
from sympy import *
from IPython.display import display, Math, Latex
init_printing()
```

## 2D

```python
vertices = np.array(symbols('v_x v_y'))
translation = np.array(symbols('w_x w_y'))
theta = symbols('\\theta')
w = [translation[0], translation[1], theta]
```

```python
R = np.array([[cos(theta), -sin(theta)], [sin(theta), cos(theta)]])
```

```python
t = R.dot(vertices) + translation
Matrix(t)
```

```python
def grad(x, w):
    return [diff(x, wi) for wi in w]
```

```python
grad_t =[grad(ti, w) for ti in t]
Matrix(grad_t)
```

```python
def hess(X, w):
    return [[diff(x, wi) for wi in w] for x in X]
```

```python
hessian_t = [hess(gti, w) for gti in grad_t]
for row in hessian_t:
    display(Matrix(row))
```

## 3D

```python
vertices = np.array(symbols('v_x v_y v_z'))
translation = np.array(symbols('w_x w_y w_z'))
thetas = np.array(symbols('\\theta_x \\theta_y \\theta_z'))
w = np.array([*translation, *thetas])
display(Matrix(w.tolist()))
```

```python
R_x = np.array([[1, 0, 0], [0, cos(thetas[0]), -sin(theta[0])], [0, sin(thetas[0]), cos(theta[0])]])
display(Matrix(R_x.tolist()))
R_y = np.array([[cos(thetas[1]), 0, sin(thetas[1])], [0, 1, 0], [-sin(thetas[1]), 0, cos(theta[1])]])
display(Matrix(R_y.tolist()))
R_z = np.array([[cos(thetas[2]), -sin(thetas[2]), 0], [sin(thetas[2]), cos(theta[2]), 0], [0, 0, 1]])
display(Matrix(R_z.tolist()))
```

```python
t = R_z @ R_y @ R_x @ vertices + translation
Matrix(t)
```

```python
grad_t =[grad(ti, w) for ti in t]
Matrix(grad_t)
```

```python
hessian_t = [hess(gti, w) for gti in grad_t]
for row in hessian_t:
    display(Matrix(row))
```
